# Create a table named install_stats which has a single column, Installs, which containg the mean number of install of apps in each type of content rating



```python
df.info()
```

```python
df["Installs"]=df["Installs"].replace({"Free":"0"})
```

# User defined functions¶

```python
def clean_data(st):
    st=st.replace(",","").replace("+","")
    return int(st)
```

```python
df["Installs"]=df["Installs"].apply(clean_data)
```

```python
df["Installs"]
```

```python
#### Examples on User Defined function an lambda functions
```

```python
df["Installs"]=df["Installs"].apply(lambda st: int(st.replace(",","").replace("+","")))
```

```python
df["Installs"]
```

```python
install_stats=df.groupby("Content Rating")["Installs"].mean() 
```

```python
install_stats
```

# Suppose a good app is one with at least one million installs and a rating of at least 4.0 Create a variable called best_category containing the name of the category with the best apps


```python
best_category=df[(df["Installs"]>=1000000) & (df["Rating"]>=4)]
best_category
```

# Make a bar chart of the average rating by category in which the bars are sorted from smallest to largest


```python
df.groupby('Category')['Rating'].mean().sort_values(ascending=True).plot(kind='bar',title='Average Rating/Category', fontsize=10)
fig = plt.figure(figsize=(20,10)) 


```
